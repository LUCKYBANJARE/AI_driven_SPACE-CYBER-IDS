```python
# ================================
# COMPLETE CICIDS CLEAN PIPELINE
# ================================

import glob
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler, LabelEncoder

# -------------------------------
# 1️⃣ Load All CICIDS CSV Files
# -------------------------------

cic_files = glob.glob("../datasets/cicids-dataset/*.csv")

if len(cic_files) == 0:
    raise FileNotFoundError("No CSV files found in cicids-dataset/ folder.")

print(f"Found {len(cic_files)} files")

df_list = []

for file in cic_files:
    print(f"Loading: {file}")
    
    df = pd.read_csv(file, low_memory=False)
    
    # Strip column whitespace
    df.columns = df.columns.str.strip()
    
    # Drop duplicate rows
    df = df.drop_duplicates()
    
    # Remove invalid negative Flow Duration
    if 'Flow Duration' in df.columns:
        df = df[df['Flow Duration'] >= 0]
    
    df_list.append(df)

# -------------------------------
# 2️⃣ Merge All Files
# -------------------------------

cic_combined = pd.concat(df_list, ignore_index=True)

print("\nFinal Combined Shape:", cic_combined.shape)

# -------------------------------
# 3️⃣ Handle Missing / Infinite Values
# -------------------------------

# Replace inf with NaN
cic_combined.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with any NaN
cic_combined.dropna(inplace=True)

print("After cleaning NaN/Inf:", cic_combined.shape)

# -------------------------------
# 4️⃣ Encode Labels
# -------------------------------

if 'Label' not in cic_combined.columns:
    raise KeyError("Label column not found.")

label_encoder = LabelEncoder()
cic_combined['Label_Encoded'] = label_encoder.fit_transform(cic_combined['Label'])

print("\nLabel Distribution:")
print(cic_combined['Label'].value_counts())

# -------------------------------
# 5️⃣ Separate Features & Target
# -------------------------------

X = cic_combined.drop(['Label', 'Label_Encoded'], axis=1)
y = cic_combined['Label_Encoded']

# Keep only numeric columns
X = X.select_dtypes(include=[np.number])

print("\nFeature Matrix Shape:", X.shape)

# -------------------------------
# 6️⃣ Robust Scaling (handles outliers)
# -------------------------------

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------
# 7️⃣ Create Stable Cyber Risk Score
# -------------------------------

# Remove zero-variance columns
variance = X_scaled.var(axis=0)
non_zero_var_cols = variance > 1e-10
X_filtered = X_scaled[:, non_zero_var_cols]

print("Removed", np.sum(~non_zero_var_cols), "zero-variance columns")

# Compute anomaly score using L2 norm (much more stable)
row_anomaly_score = np.linalg.norm(X_filtered, axis=1)

# Normalize safely
min_val = np.min(row_anomaly_score)
max_val = np.max(row_anomaly_score)

if max_val - min_val == 0:
    cyber_risk_score = np.zeros_like(row_anomaly_score)
else:
    cyber_risk_score = (row_anomaly_score - min_val) / (max_val - min_val)

cic_combined['Cyber_Risk_Score'] = cyber_risk_score

print("\nCyber Risk Score Created Successfully")
print("Risk Score Range:",
      cic_combined['Cyber_Risk_Score'].min(),
      "to",
      cic_combined['Cyber_Risk_Score'].max())
# -------------------------------
# 8️⃣ Save Cleaned Dataset
# -------------------------------

cic_combined.to_csv("cicids_cleaned_with_risk.csv", index=False)

print("\nSaved: cicids_cleaned_with_risk.csv")

# -------------------------------
# 9️⃣ Final Summary
# -------------------------------

print("\n===== FINAL SUMMARY =====")
print("Rows:", cic_combined.shape[0])
print("Columns:", cic_combined.shape[1])
print("Risk Score Range:",
      cic_combined['Cyber_Risk_Score'].min(),
      "to",
      cic_combined['Cyber_Risk_Score'].max())
```

```python
def propagate_satellite(tle1, tle2, minutes_ahead=60):
    sat = Satrec.twoline2rv(tle1, tle2)

    future_time = datetime.utcnow() + timedelta(minutes=minutes_ahead)
    jd, fr = jday(future_time.year, future_time.month, future_time.day,
                  future_time.hour, future_time.minute, future_time.second)

    e, r, v = sat.sgp4(jd, fr)

    return np.array(r), np.array(v)
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# =========================
# 1. Load
# =========================
train = pd.read_csv("../datasets/EAS_train_data.csv")
test = pd.read_csv("../datasets/EAS_test_data.csv")

# =========================
# 2. Clean
# =========================
train = train.replace([np.inf, -np.inf], np.nan)
test = test.replace([np.inf, -np.inf], np.nan)

train = train.dropna()
test = test.dropna()

# =========================
# 3. Feature Engineering
# =========================

def add_features(df):
    
    df['velocity_mag'] = np.sqrt(
        df['relative_velocity_r']**2 +
        df['relative_velocity_t']**2 +
        df['relative_velocity_n']**2
    )
    
    df['position_mag'] = np.sqrt(
        df['relative_position_r']**2 +
        df['relative_position_t']**2 +
        df['relative_position_n']**2
    )
    
    df['probability'] = np.exp(df['risk'])
    
    df['congestion_index'] = (
        (1 / (df['miss_distance'] + 1e-6)) *
        df['velocity_mag'] *
        df['mahalanobis_distance'] *
        df['probability']
    )
    
    return df

train = add_features(train)
test = add_features(test)

# =========================
# 4. Select Congestion Features
# =========================

features = [
    'velocity_mag',
    'position_mag',
    'miss_distance',
    'mahalanobis_distance',
    'time_to_tca',
    'F10', 'F3M', 'SSN', 'AP',
    't_j2k_sma', 'c_j2k_sma',
    't_j2k_ecc', 'c_j2k_ecc',
    't_j2k_inc', 'c_j2k_inc'
]

X_train = train[features]
y_train = train['congestion_index']

X_test = test[features]
y_test = test['congestion_index']

# =========================
# 5. Train Forecast Model
# =========================

model = XGBRegressor(
    n_estimators=800,
    max_depth=8,
    learning_rate=0.03,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Congestion Forecast MSE:",
      mean_squared_error(y_test, pred))
```

```python
# ================================
# AUTONOMOUS MANEUVER RECOMMENDATION
# ================================

import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, classification_report

# =========================
# 1. Load Data
# =========================
train = pd.read_csv("../datasets/EAS_train_data.csv")
test = pd.read_csv("../datasets/EAS_test_data.csv")

# =========================
# 2. Clean
# =========================
train = train.replace([np.inf, -np.inf], np.nan).dropna()
test = test.replace([np.inf, -np.inf], np.nan).dropna()

# =========================
# 3. Feature Engineering
# =========================
def add_features(df):

    df['velocity_mag'] = np.sqrt(
        df['relative_velocity_r']**2 +
        df['relative_velocity_t']**2 +
        df['relative_velocity_n']**2
    )

    df['position_mag'] = np.sqrt(
        df['relative_position_r']**2 +
        df['relative_position_t']**2 +
        df['relative_position_n']**2
    )

    df['probability'] = np.exp(df['risk'])

    df['congestion_index'] = (
        (1 / (df['miss_distance'] + 1e-6)) *
        df['velocity_mag'] *
        df['mahalanobis_distance'] *
        df['probability']
    )

    return df

train = add_features(train)
test = add_features(test)

# =========================
# 4. Define Maneuver Threshold
# =========================
threshold = train['congestion_index'].quantile(0.95)

train['maneuver_required'] = (
    train['congestion_index'] > threshold
).astype(int)

test['maneuver_required'] = (
    test['congestion_index'] > threshold
).astype(int)

# =========================
# 5. Select Features
# =========================
features = [
    'velocity_mag',
    'position_mag',
    'miss_distance',
    'mahalanobis_distance',
    'time_to_tca',
    'F10', 'F3M', 'SSN', 'AP',
    't_j2k_sma', 'c_j2k_sma',
    't_j2k_ecc', 'c_j2k_ecc',
    't_j2k_inc', 'c_j2k_inc'
]

X_train = train[features]
y_train = train['maneuver_required']

X_test = test[features]
y_test = test['maneuver_required']

# =========================
# 6. Handle Class Imbalance
# =========================
scale_weight = (len(y_train) - y_train.sum()) / y_train.sum()

# =========================
# 7. Train Model
# =========================
model = XGBClassifier(
    n_estimators=700,
    max_depth=7,
    learning_rate=0.03,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_weight,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)

# =========================
# 8. Evaluate
# =========================
proba = model.predict_proba(X_test)[:, 1]
pred = (proba > 0.5).astype(int)

print("Maneuver AUC:", roc_auc_score(y_test, proba))
print(classification_report(y_test, pred))
```
